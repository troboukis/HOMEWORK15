# Grouped bar graph (2 colors)

```python
df = pd.read_csv("gender.csv")
df.head(3)
```

```python
ax = df.plot(x='state', kind='barh', figsize=(10,4))

# CUSTOMIZATION CODE GOES UNDER HERE
ax.title.set_fontname('Silom')
ax.set_title('Population by Gender in the United States', loc='left')
ax.title.set_fontsize(22)
ax.title.set_fontweight(weight='black')
ax.set(ylabel="")

ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['right'].set_visible(False)
ax.tick_params(which='major', left=False, bottom=False)

ax.grid(True, axis='y', color='lightgray', linestyle='')
ax.grid(True, axis='x', color='lightgray', linestyle='-')

# then we save it
plt.savefig("graph-bar-grouped2.pdf")
plt.show()
```

# Bar graph (Single color)

```python
df = pd.read_csv("countries.csv")
result = df.groupby("continent").life_expectancy.median().sort_values()
result
```

```python
color = plt.rcParams['axes.prop_cycle'].by_key()['color'][1]
ax = result.plot(kind='barh', figsize=(10, 5), color=color)

# CUSTOMIZATION CODE GOES UNDER HERE
ax.title.set_fontname('Silom')
ax.set_title('Media Life Expectancy by Contintent', loc='left')
ax.title.set_fontsize(22)
ax.title.set_fontweight(weight='black')
ax.set(ylabel="")

ax.spines['top'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.tick_params(which='major', left=False, bottom=False)

ax.grid(True, axis='y', color='lightgray', linestyle='')
ax.grid(True, axis='x', color='lightgray', linestyle='-')

# then we save it
plt.savefig("graph-bar3.pdf")
plt.show()
```

# Line graph

```python
df = pd.read_csv("visas-by-continent-long.csv")
df = df.groupby(["year", "continent"]).visas.sum().unstack().reset_index()
df.drop(columns='Unknown', inplace=True)
df['year'] = pd.to_datetime(df.year, format="%Y")
df.set_index('year', inplace=True)
percent_df = df.div(df.sum(axis=1), axis='rows') * 100

percent_df.head(2)
```

```python
ax = percent_df.plot(kind='line', figsize=(10, 5))

# CUSTOMIZATION CODE GOES UNDER HERE
ax.title.set_fontname('Silom')
ax.set_title('Visas Issued by Contintent', loc='left')
ax.title.set_fontsize(22)
ax.title.set_fontweight(weight='black')
ax.set(ylabel="")

ax.spines['top'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.tick_params(which='major', left=False, bottom=False)

ax.grid(True, axis='y', color='lightgray', linestyle='-')
ax.grid(True, axis='x', color='lightgray', linestyle='')

# then we save it
plt.savefig("graph-lines3.pdf")
plt.show()
```

# Scatter plot

```python
df = pd.read_csv("countries.csv")
df.head(2)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Don't touch this part of the code!
colors = list(reversed(plt.rcParams['axes.prop_cycle'].by_key()['color']))

groups = df.groupby('continent')
for name, group in groups:
    group.plot(kind='scatter', x='gdp_per_capita', y='life_expectancy', color=colors.pop(), label=name, s= 5 + (group.population / 200000), ax=ax, alpha=0.75)
ax.legend_.remove()

# CUSTOMIZATION CODE GOES UNDER HERE

ax.title.set_fontname('Silom')
ax.set_title('How Income Affects Life Expectancy', loc='left')
ax.title.set_fontsize(22)
ax.title.set_fontweight(weight='black')
ax.set(ylabel="")

ax.spines['top'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.tick_params(which='major', left=False, bottom=False)

ax.grid(True, axis='y', color='lightgray', linestyle='-')
ax.grid(True, axis='x', color='lightgray', linestyle='')

# then we save it
plt.savefig("graph-scatter3.pdf")
plt.show()
```
